```python
trd_ctx = OpenSecTradeContext(host='127.0.0.1', port=11111)

# Get account list
ret, data = trd_ctx.get_acc_list()
if ret == RET_OK:
    print("Account List:")
    print(data)
else:
    print('Error:', data)

# Get positions for paper trading account
ret, data = trd_ctx.position_list_query(trd_env=TrdEnv.SIMULATE)
if ret == RET_OK:
    print("Positions:")
    print(data)
else:
    print('Error:', data)

# Get account funds/cash info
ret, data = trd_ctx.accinfo_query(trd_env=TrdEnv.SIMULATE)
if ret == RET_OK:
    print("Account Info:")
    print(data)
else:
    print('Error:', data)

trd_ctx.close()
```

```python
def get_stock_price(ticker):
    quote_ctx = OpenQuoteContext(host='127.0.0.1', port=11111)
    
    try:
        # Get current price
        ret, data = quote_ctx.get_market_snapshot([ticker])
        if ret == RET_OK:
            stock_data = data.iloc[0]
            print(f"Ticker: {stock_data['code']}")
            print(f"Current Price: ${stock_data['last_price']}")
            print(f"Open: ${stock_data['open_price']}")
            print(f"High: ${stock_data['high_price']}")
            print(f"Low: ${stock_data['low_price']}")
            print(f"Volume: {stock_data['volume']}")
            return stock_data['last_price']
        else:
            print(f"Error getting data: {data}")
            return None
    finally:
        quote_ctx.close()
```

```python
get_stock_price('US.IBM')
```
